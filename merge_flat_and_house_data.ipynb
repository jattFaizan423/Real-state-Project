```python
import numpy as np
import pandas as pd
```

```python
flats = pd.read_csv('/content/flats_cleaned.csv')
houses = pd.read_csv('/content/house_cleaned.csv')
```

```python
df = pd.concat([flats,houses],ignore_index=True)
```

```python
df = df.sample(df.shape[0],ignore_index=True)
```
